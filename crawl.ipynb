```python
#Fungsi pembersihan data
def cleanText(text):
    text = text.lower()
    text = re.sub(r'(@[A-Za-z0-9_]+)', '', text)
    text = re.sub('http://\S+|https://\S+', '', text)
    text = re.sub(r'[^\w\s]', '', text)

    text = ''.join(text)
    return text
```

```python
# Penentuan polaritas sentimen
def sentiment(cleaned_text):
  if cleaned_text.sentiment.polarity > 0:
    return 'positive'
  elif cleaned_text.sentiment.polarity < 0:
    return 'negative'
  else:
    return 'neutral'
```

```python
# Pengambilan data dari twitter
tweets = [] 
all_polarity = 0
for tweet in tweepy.Cursor(method=api.search_tweets, q="Digital pollution -filter:retweets", tweet_mode='extended').items(100):
    # Text Cleaning
    txt =  tweet.full_text
    cleanTxt = cleanText(txt)
    #Sentiment Analysis
    analysis = TextBlob(cleanTxt)
    try:
        an = analysis.translate(from_lang='en', to='en')
        sent = sentiment(an)
    except:
        sent = sentiment(analysis)
    tweets.append((tweet.created_at, cleanTxt, tweet.retweet_count, tweet.favorite_count, sent))     
    tweetsdf=pd.DataFrame(tweets, columns=["Tanggal Tweet","Tweet Text","Jumlah Retweet","Jumlah Likes", "Sentiment"])
```
